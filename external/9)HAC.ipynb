```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_excel("/content/hierarchical_clustering (1).xlsx")
data
```

```python
x = data["Variable 1"]
y = data["Variable 2"]
plt.scatter(x, y, marker = 'o', c = 'red' )
plt.grid()
plt.xlabel('Variable_1')
plt.ylabel('Variable_2')
```

```python
linked = linkage(data,'single')
labelList = range(1,8)

dendrogram(linked,labels=labelList)
plt.show()
```

```python
Hclustering = AgglomerativeClustering(n_clusters = None, linkage='single',distance_threshold=3)
Hclustering.fit(data)
```

```python
result = Hclustering.fit_predict(data)
result
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c = Hclustering.labels_, cmap ='rainbow')
plt.grid()
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
```

```python
silhouette_score(data, result)
```
