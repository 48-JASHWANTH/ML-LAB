```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel('/content/TRUCKING.xlsx')
data
```

```python
plt.scatter(data['miles_travelled'], data['travel_time'])
plt.title('trend between indpendent and dependent variable')
plt.xlabel = ('miles travelled')
plt.ylabel = ('travel time')
plt.show()
```

```python
x = data[['miles_travelled', 'n_of_deliveries']]
y = data['travel_time']
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=45)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
y_pred= reg.predict(x_test)
y_pred
```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print(reg.score(x_train, y_train))
```
